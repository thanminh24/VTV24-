```python
import requests

def get_channel_playlists(channel_id, api_key):
    """
    Retrieves all playlists associated with a YouTube channel.

    Args:
        channel_id: The ID of the YouTube channel.
        api_key: Your YouTube Data API key.

    Returns:
        A list of playlist IDs.
    """

    playlist_ids = []
    next_page_token = None
    base_url = "https://www.googleapis.com/youtube/v3/playlists"

    while True:
        params = {
            "part": "snippet",
            "channelId": channel_id,
            "key": api_key,
            "maxResults": "50"
        }
        if next_page_token:
            params["pageToken"] = next_page_token
        url = base_url + "?" + "&".join(f"{k}={v}" for k, v in params.items())

        response = requests.get(url)
        data = response.json()

        for item in data['items']:
            playlist_ids.append(item['id'])

        next_page_token = data.get('nextPageToken')
        if not next_page_token:
            break

    return playlist_ids

def get_playlist_video_ids(playlist_id, api_keys):
    """
    Retrieves all video IDs from a YouTube playlist, handling API quota errors
    by switching keys.

    Args:
        playlist_id: The ID of the YouTube playlist.
        api_keys: A list of YouTube Data API keys.

    Returns:
        A list of video IDs.
    """

    video_ids = []
    next_page_token = None
    base_url = "https://www.googleapis.com/youtube/v3/playlistItems"
    current_key_index = 0

    while True:
        if current_key_index >= len(api_keys):
            print("All API keys have been exhausted.")
            break

        params = {
            "part": "snippet",
            "playlistId": playlist_id,
            "key": api_keys[current_key_index],
            "maxResults": "50"
        }
        if next_page_token:
            params["pageToken"] = next_page_token
        url = base_url + "?" + "&".join(f"{k}={v}" for k, v in params.items())

        try:
            response = requests.get(url)
            data = response.json()

            for item in data['items']:
                video_ids.append(item['snippet']['resourceId']['videoId'])

            next_page_token = data.get('nextPageToken')
            if not next_page_token:
                break

        except KeyError as e:
            if "quotaExceeded" in str(e):
                print(f"API key {api_keys[current_key_index]} quota exceeded. Switching to the next key...")
                current_key_index += 1
            else:
                print(f"An unexpected KeyError occurred: {e}")
                break
        except Exception as e:
            print(f"An error occurred: {e}")
            break

    return video_ids


# Example usage with your provided channel ID and API keys
channel_id = "UCabsTV34JwALXKGMqHpvUiA"
api_key = "AIzaSyBBTi28wWqW7TaNSXM_BD0qm_K0JMU4qaE"  # Use a single API key for get_channel_playlists
api_keys = [ 'AIzaSyBB66rRVXWvzGp4wFUMEcFEGAchCoeJI3g','AIzaSyCHMr5otVsepfxk3a89uA4zh0fe_6K8nWk','AIzaSyAmVGsZrA7tgaGgzLFn_C6w7llhqMZMWkM','AIzaSyCPApuMx3wyGsby0_A40exdyyrjZU6bvq4' ]

try:
    playlist_ids = get_channel_playlists(channel_id, api_key)
    print("Playlist IDs:", playlist_ids)
    print("Total Playlists:", len(playlist_ids))

    all_video_ids = []
    for playlist_id in playlist_ids:
        video_ids = get_playlist_video_ids(playlist_id, api_keys)
        print(f"Video IDs for playlist {playlist_id}: {video_ids}")
        all_video_ids.extend(video_ids)

    print("\nAll Video IDs:", all_video_ids)
    print("Total Videos:", len(all_video_ids))

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import pandas as pd
from googleapiclient.discovery import build

# List of your YouTube Data API keys
API_KEYS = ['AIzaSyBITsCwh_7xUaI92ebhtEVjSiGmerHrgdE', 'AIzaSyBuC3ZFPFZL36y35gpH_vIYjhtZXzd9WOU', 'AIzaSyAMYdnedg-A3mReLIJ6gqM1suzLmjUZjIY','AIzaSyCFDO6vhQlv1b1T-ioyrptn2segRZ_3PkQ','AIzaSyBU8m2RoSRWiTl4qYGV1tl172R6W5MRv9c']  # Replace with your actual API keys

# Initialize the current API key index
current_key_index = 0

def get_youtube_service():
    """
    Returns a YouTube Data API v3 service object with the current API key.
    """
    global current_key_index
    return build("youtube", "v3", developerKey=API_KEYS[current_key_index])

def get_video_stats(video_id):
    """
    Retrieves statistics for a given YouTube video.

    Args:
        video_id: The ID of the YouTube video.

    Returns:
        A dictionary containing video statistics or None if the video is not found.
    """
    global current_key_index
    youtube = get_youtube_service()

    try:
        response = youtube.videos().list(
            part="snippet,statistics",
            id=video_id
        ).execute()

        if "items" in response and response["items"]:
            video_data = response["items"][0]
            snippet = video_data["snippet"]
            statistics = video_data["statistics"]
            print(f"Video found: {video_id}")
            return {
                "video_id": video_id,
                "published_date": snippet["publishedAt"],
                "video_title": snippet["title"],
                "video_description": snippet["description"],
                "likes": int(statistics.get("likeCount", 0)),
                "views": int(statistics["viewCount"])
            }
        else:
            print(f"Video not found: {video_id}")
            return None

    except Exception as e:
        print(f"Error retrieving video stats: {e}")
        if "quotaExceeded" in str(e):  # Check for quota exceeded error
            print("API key quota exceeded.")
            current_key_index = (current_key_index + 1) % len(API_KEYS)
            if current_key_index == 0:  # All keys exhausted
                print("All API keys exhausted. Returning None.")
                return None
            else:
                print(f"Switching to API key: {API_KEYS[current_key_index]}")
                return get_video_stats(video_id)  # Retry with the new key
        else:
            return None


def create_video_stats_df(video_ids):
    """
    Creates a Pandas DataFrame with statistics for multiple YouTube videos.

    Args:
        video_ids: A list of YouTube video IDs.

    Returns:
        A Pandas DataFrame with video statistics.
    """
    video_stats_data = []
    for video_id in video_ids:
        stats = get_video_stats(video_id)
        if stats:
            video_stats_data.append(stats)

    df = pd.DataFrame(video_stats_data,
                      columns=['video_id', 'published_date', 'video_title',
                               'video_description', 'likes', 'views'])
    return df


# Example usage:
  # Replace with actual video IDs
df = create_video_stats_df(all_video_ids)
df
```
